```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Qiskit imports (Qiskit 2.x / qiskit-aer)
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError
from qiskit.quantum_info import Operator

# ---------------------------
# Helper plotting function
# ---------------------------
def plot_counts(counts, title="Counts", top_n=10):
    """Simple bar plot for counts (dict). Shows up to top_n keys sorted by counts."""
    if not counts:
        print("No counts to plot.")
        return
    items = sorted(counts.items(), key=lambda kv: kv[1], reverse=True)[:top_n]
    labels = [k for k, _ in items]
    vals = [v for _, v in items]
    plt.figure(figsize=(max(6, 0.6 * len(labels)), 3.5))
    plt.bar(labels, vals)
    plt.title(title)
    plt.xlabel("bitstring (classical register)")
    plt.ylabel("counts")
    plt.show()

# ---------------------------
# Oracle: f(x) = s·x ⊕ b
# ---------------------------
def bv_oracle(qc, input_qubits, ancilla, s_bits, b_bit=0):
    # apply constant bit b to ancilla (affects ancilla only)
    if int(b_bit) == 1:
        qc.x(ancilla)
    # apply CNOTs according to s bits
    for i, sbit in enumerate(s_bits):
        if int(sbit) == 1:
            qc.cx(input_qubits[i], ancilla)

# ---------------------------
# Build Bernstein–Vazirani circuits
# ---------------------------
def bv_circuit(n, s_bits, b_bit=0, measure_ancilla=False):
    qr_in = QuantumRegister(n, 'x')
    qr_anc = QuantumRegister(1, 'anc')
    cr_in = ClassicalRegister(n, 'c_in')
    qc = QuantumCircuit(qr_in, qr_anc, cr_in)

    # initialize ancilla to |1>
    qc.x(qr_anc)
    # Hadamards: ancilla and inputs
    qc.h(qr_anc)
    qc.h(qr_in)

    # oracle
    bv_oracle(qc, qr_in, qr_anc[0], s_bits, b_bit)

    # Hadamards on inputs again
    qc.h(qr_in)

    # measure inputs
    qc.measure(qr_in, cr_in)

    # optional ancilla measurement (if requested, user can create a wrapper that adds ancilla classical bit)
    if measure_ancilla:
        # we need to rebuild circuit to include ancilla classical bit
        cr_anc = ClassicalRegister(1, 'c_anc')
        qc.add_register(cr_anc)
        qc.measure(qr_anc, cr_anc)

    return qc

def bv_circuit_with_ancilla_measure(n, s_bits, b_bit=0):
    """Convenience: explicit ancilla measurement returning a circuit with two classical registers"""
    qr_in = QuantumRegister(n, 'x')
    qr_anc = QuantumRegister(1, 'anc')
    cr_in = ClassicalRegister(n, 'c_in')
    cr_anc = ClassicalRegister(1, 'c_anc')
    qc = QuantumCircuit(qr_in, qr_anc, cr_in, cr_anc)

    qc.x(qr_anc)
    qc.h(qr_anc)
    qc.h(qr_in)

    bv_oracle(qc, qr_in, qr_anc[0], s_bits, b_bit)

    qc.h(qr_in)
    qc.measure(qr_in, cr_in)
    qc.measure(qr_anc, cr_anc)
    return qc

# ---------------------------
# Utilities for comparing classical outcomes to s (bit-order handling)
# ---------------------------
def most_likely_bitstring(counts):
    """Return the key with the highest count (or None)."""
    if not counts:
        return None
    return max(counts.items(), key=lambda kv: kv[1])[0]

def canonicalize_key(key, n):
    # return both possibilities (normal, reversed)
    return key, key[::-1]

def compare_to_secret(measured_key, s):
    if measured_key is None:
        return False, None, None
    normal, rev = measured_key, measured_key[::-1]
    if normal == s:
        return True, 'normal', normal
    if rev == s:
        return True, 'reversed', rev
    return False, None, None

# ---------------------------
# Simulation runners (ideal and noisy)
# ---------------------------
def run_ideal(qc, shots=1024):
    backend = AerSimulator()
    tcirc = transpile(qc, backend=backend)
    job = backend.run(tcirc, shots=shots)
    result = job.result()
    counts = result.get_counts()
    return counts

def run_noisy(qc, noise_model, shots=2048):
    """
    Run circuit on AerSimulator with provided noise_model (qiskit_aer.noise.NoiseModel).
    """
    backend = AerSimulator()
    tcirc = transpile(qc, backend=backend)
    job = backend.run(tcirc, shots=shots, noise_model=noise_model)
    res = job.result()
    return res.get_counts()

# ---------------------------
# Build a sample NoiseModel
# ---------------------------
def make_simple_noise_model(p1=0.002, p2=0.01, readout_err=0.03):
    nm = NoiseModel()
    single_err = depolarizing_error(p1, 1)
    two_err = depolarizing_error(p2, 2)
    # Add single-qubit errors to common single-qubit gates
    nm.add_all_qubit_quantum_error(single_err, ['u3', 'u2', 'u1', 'h', 'x'])
    nm.add_all_qubit_quantum_error(two_err, ['cx'])
    # Readout error
    ro = ReadoutError([[1 - readout_err, readout_err], [readout_err, 1 - readout_err]])
    nm.add_all_qubit_readout_error(ro)
    return nm

# ---------------------------
# Main demo: change s and b, verify measured outputs
# ---------------------------
def demo(secret_s, b, shots_ideal=1024, shots_noisy=2048, plot=True):
    """
    secret_s: string like '10110' (length n)
    b: integer 0 or 1
    """
    n = len(secret_s)
    print(f"\n--- Bernstein–Vazirani demo: n={n}, s={secret_s}, b={b} ---")

    # Build circuits
    qc = bv_circuit(n, secret_s, b_bit=b, measure_ancilla=False)
    qc_with_anc = bv_circuit_with_ancilla_measure(n, secret_s, b_bit=b)

    print("\nCircuit (compact):")
    print(qc.draw(fold=120))

    # 1) ideal run (inputs only)
    counts_ideal = run_ideal(qc, shots=shots_ideal)
    print("\nIdeal simulator counts (inputs only):")
    print(counts_ideal)
    ml = most_likely_bitstring(counts_ideal)
    match, orientation, normalized = compare_to_secret(ml, secret_s)
    if match:
        print(f"Measured most-likely bitstring '{ml}' matches secret s when using orientation='{orientation}' (interpreted as '{normalized}')")
    else:
        print(f"Most-likely measured '{ml}' does NOT match secret s '{secret_s}' (try checking bit-order).")

    if plot:
        plot_counts(counts_ideal, title=f"Ideal counts (n={n}, s={secret_s}, b={b})")

    # 2) measure ancilla as well (ideal) to inspect b
    counts_ideal2 = run_ideal(qc_with_anc, shots=shots_ideal)
    print("\nIdeal (inputs + ancilla) counts (format: 'inputbits ancilla'):")
    # Note: when measuring two classical registers, the returned keys format is 'ancilla_bit input_bits'
    # depending on Qiskit version; we'll print some keys to see ordering.
    print(dict(list(counts_ideal2.items())[:10]))
    if plot:
        plot_counts(counts_ideal2, title="Ideal (inputs+ancilla) counts")

    # Find ancilla statistics by splitting keys; handle potential orders:
    # Determine typical key format by checking lengths
    any_key = next(iter(counts_ideal2))
    # When cr_anc is last register in circuit, key is 'ancilla_bit inputbits' (ancilla first), or reversed.
    # We detect by lengths:
    # If len(key) == n+1, then either ancilla is leftmost or rightmost
    anc_counts = {}
    for k, v in counts_ideal2.items():
        if len(k) == n + 1:
            # try both splits
            left = k[0]      # ancilla candidate
            right = k[1:]    # input candidate
            # decide which side corresponds to ancilla by checking whether right matches secret (or reversed)
            if right == secret_s or right[::-1] == secret_s:
                anc = left
            else:
                # maybe ancilla is rightmost
                left2 = k[:-1]
                right2 = k[-1]
                if left2 == secret_s or left2[::-1] == secret_s:
                    anc = right2
                else:
                    # fallback: assume ancilla is last char
                    anc = k[-1]
            anc_counts[anc] = anc_counts.get(anc, 0) + v
        else:
            # unexpected: just categorize by last bit
            anc = k[-1]
            anc_counts[anc] = anc_counts.get(anc, 0) + v
    print("Aggregated ancilla counts (ideal):", anc_counts)
    if plot:
        plot_counts(anc_counts, title="Ancilla distribution (ideal aggregated)")

    # 3) Noisy run (inputs only) using a simple noise model
    noise_model = make_simple_noise_model(p1=0.005, p2=0.02, readout_err=0.03)
    counts_noisy = run_noisy(qc, noise_model, shots=shots_noisy)
    print("\nNoisy simulator counts (inputs only):")
    print(counts_noisy)
    ml_noisy = most_likely_bitstring(counts_noisy)
    match_noisy, orientation_noisy, normalized_noisy = compare_to_secret(ml_noisy, secret_s)
    if match_noisy:
        print(f"Noisy most-likely bitstring '{ml_noisy}' matches secret s when using orientation='{orientation_noisy}' (interpreted as '{normalized_noisy}')")
    else:
        print(f"Noisy most-likely measured '{ml_noisy}' does NOT match secret s '{secret_s}'. Noise reduced reliability.")
    if plot:
        plot_counts(counts_noisy, title=f"Noisy counts (n={n}, s={secret_s}, b={b})")

    # 4) Noisy run with ancilla measure to inspect effect of b under noise
    counts_noisy2 = run_noisy(qc_with_anc, noise_model, shots=shots_noisy)
    print("\nNoisy (inputs+ancilla) counts (sample):")
    print(dict(list(counts_noisy2.items())[:10]))
    # aggregate ancilla similarly
    anc_counts_noisy = {}
    for k, v in counts_noisy2.items():
        if len(k) == n + 1:
            # try to guess anc bit as first or last char; simpler approach: tab both possibilities
            # We'll aggregate by both possible anc positions to show the effect.
            anc_first = k[0]
            anc_last = k[-1]
            # count both separately using keys 'first:bit' and 'last:bit' (for debugging)
            anc_counts_noisy.setdefault('first_' + anc_first, 0)
            anc_counts_noisy['first_' + anc_first] += v
            anc_counts_noisy.setdefault('last_' + anc_last, 0)
            anc_counts_noisy['last_' + anc_last] += v
        else:
            anc = k[-1]
            anc_counts_noisy.setdefault('last_' + anc, 0)
            anc_counts_noisy['last_' + anc] += v
    print("Aggregated ancilla counts (noisy candidates):", anc_counts_noisy)
    if plot:
        plot_counts(anc_counts_noisy, title="Ancilla distribution (noisy aggregated candidates)")

    # 5) Print oracle unitary for very small n (n <= 3 advisable)
    if n <= 3:
        oracle_only = QuantumCircuit(n + 1)
        bv_oracle(oracle_only, oracle_only.qubits[:n], oracle_only.qubits[n], secret_s, b_bit=b)
        op = Operator(oracle_only)
        np.set_printoptions(precision=3, suppress=True)
        print("\nOracle unitary matrix (dense):")
        print(op.data)

    # Return dictionaries for possible further processing
    return {
        'counts_ideal': counts_ideal,
        'counts_ideal_with_anc': counts_ideal2,
        'counts_noisy': counts_noisy,
        'counts_noisy_with_anc': counts_noisy2,
        'anc_ideal_aggregated': anc_counts,
        'anc_noisy_aggregated': anc_counts_noisy,
    }


# ---------------------------
# Example runs (change s and b here)
# ---------------------------
if __name__ == "__main__":
    # Examples to test multiple secrets and b values
    examples = [
        ('101', 0),
        ('101', 1),
        ('10110', 1),
        ('01011', 0),
    ]
    for s, b in examples:
        out = demo(s, b, shots_ideal=1024, shots_noisy=2048, plot=True)
```
